```python
# lets reanme the index and colunm names by user.id and book title
books_df= books_title_d.pivot (index="User.ID",
                     columns="Book.Title",
                     values="Book.Rating").reset_index(drop=True)
books_df
```

```python
# filling the null values by zeros
books_df1=books_df.fillna(0, axis=1)
books_df1
```

```python
# lets check for the similairities using cosine function
from scipy.spatial.distance import cosine
from sklearn.metrics import pairwise_distances
import scipy
```

```python
books_similiar= 1- pairwise_distances(books_df1.values, metric="cosine")
books_similiar
```

```python
# forming the dataframe for the books similiarity
books_similiar_df=pd.DataFrame(books_similiar)
books_similiar_df
```

```python
books_similiar_df1=books_similiar_df.iloc[0:10,0:10]
```

```python
books_similiar_df1
```

```python
# lets check for the recommendation
def recommend(users_id):
    temp= list(books_similiar_df.sort_values([users_id],ascending=True).head(1000).index)
    books_list=[]
    for i in temp:
        books_list=books_list+list(books1[books1["User.ID"]==i]["Book.Title"])
    return set(books_list)-set(books1[books1["User.ID"]==users_id]["Book.Title"])
                             
```

```python
recommend(200) # for user_id 200 this is the recommended books with titles provided
```

```python
recommend(150)
```
